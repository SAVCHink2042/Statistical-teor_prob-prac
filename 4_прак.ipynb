```python
import numpy as np
from scipy import stats
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sb
n = 20

r = np.empty((1000,n))
rS = np.empty(1000)
for i in range(1000):
    r[i] = stats.uniform.rvs(size = n)
    rS[i] = np.sum(r[i])

ecdfR = stats.ecdf(rS)

mu = n * r.mean()
sigma = np.sqrt(n * r.var())

x = np.linspace(min(rS),max(rS),200)
y = np.array([ecdfR.cdf.evaluate(i) for i in x])
y1 = stats.norm.cdf(x, mu, sigma)
plt.plot(x, y)
plt.plot(x, y1)
```
